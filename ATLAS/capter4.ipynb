3. A rede no arquivo http://www.networkatlas.eu/exercises/4/3/data.txt é um hipergráfico, com uma hiperaresta por linha. Transforme-o em uma rede unipartida na qual cada hiperaresta é dividida em arestas conectando todos os nós na hiperaresta. Em seguida, transforme-o em uma rede bipartida na qual cada hiperaresta é um nó de um tipo e seus nós se conectam a ele.

```python
import networkx as nx
hyperedges = [
    [1, 2, 3, 4, 5, 6],
    [2, 4, 7, 8],
    [4, 5, 6, 7, 10],
    [6, 9, 10, 11, 12],
    [11, 13, 14, 15],
    [11, 12, 13],
    [14, 15, 16, 17, 18],
    [15, 17, 19, 20],
    [21, 22, 23],
    [23, 24],
    [24, 25, 26],
    [25, 27, 28, 29]
]

## Rede Unipartida

# Grafo unipartido
unipartite_graph = nx.Graph()

# Para cada hiperaresta, adicionar arestas entre todos os nós da hiperaresta
for edge in hyperedges:
    for i in range(len(edge)):
        for j in range(i + 1, len(edge)):
            unipartite_graph.add_edge(edge[i], edge[j])

print("Arestas da rede unipartida: ", list(unipartite_graph.edges()))

## Rede Bipartida

bipartite_graph = nx.Graph()

# Adicionar hiperarestas como nós de um tipo
for i, edge in enumerate(hyperedges):
    hyperedge_node = f"he{i+1}" # he -> hyper edge

    for node in edge:
        bipartite_graph.add_edge(hyperedge_node, node)

print("Arestas da rede bipartida: ", list(bipartite_graph.edges()))
```

4. A rede no arquivo [http://www.networkatlas.eu/exercises/4/4/data.txt](http://www.networkatlas.eu/exercises/4/4/data.txt) é dinâmica; a terceira e a quarta colunas da lista de arestas informam o primeiro e o último momento em que a aresta esteve continuamente presente. Uma aresta pode reaparecer se estiver presente em dois períodos de tempo não contínuos. Agregue-a usando uma janela disjunta de tamanho 3.

```python
import networkx as nx

# Dados das arestas com tempos (nó1, nó2, início, fim)
edges = [
    (1, 2, 1, 4),
    (1, 2, 7, 9),
    (1, 3, 3, 7),
    (1, 4, 2, 4),
    (1, 5, 1, 9),
    (2, 3, 1, 2),
    (2, 4, 1, 3),
    (2, 4, 6, 7),
    (2, 4, 9, 9),
    (3, 5, 5, 7),
    (3, 6, 1, 3),
    (3, 6, 8, 9),
    (4, 7, 1, 9),
    (5, 8, 1, 2),
    (5, 8, 4, 7),
    (5, 8, 9, 9),
    (5, 9, 1, 5),
    (5, 10, 8, 9),
    (5, 11, 7, 8),
    (6, 12, 4, 7),
    (6, 13, 2, 3),
    (7, 12, 2, 6),
]

# Tamanho da janela
window_size = 3

# Criar um dicionário para armazenar grafos por janela de tempo
graphs_by_window = {}

# Função para determinar a janela de tempo
def get_window_range(start, end, window_size):
    windows = []
    for i in range(start, end + 1):
        window_start = ((i - 1) // window_size) * window_size + 1
        window_end = window_start + window_size - 1
        if (window_start, window_end) not in windows:
            windows.append((window_start, window_end))
    return windows

# Criar os grafos para cada janela e adicionar as arestas
for u, v, start, end in edges:
    # Determinar as janelas que a aresta atravessa
    windows = get_window_range(start, end, window_size)
    
    for window_start, window_end in windows:
        # Se o grafo dessa janela ainda não existe, crie-o
        if (window_start, window_end) not in graphs_by_window:
            graphs_by_window[(window_start, window_end)] = nx.Graph()
        
        # Adicionar a aresta no grafo da janela
        graphs_by_window[(window_start, window_end)].add_edge(u, v)

# Exibir as arestas agregadas por janela
for window, graph in graphs_by_window.items():
    print(f"Janela {window}: Arestas {list(graph.edges())}")
```
